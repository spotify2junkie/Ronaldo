```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('fifa.csv')
df.head()
```

```python
candidate = df[(df['Position']=='ST')|(df['Position']=='LS')| (df['Position']=='RS')]
candidate.head()
```

```python
infor = ['Name','Nationality','Club','Age','Position','Overall','Potential','Value']
abilities = ['Crossing','Finishing','HeadingAccuracy','ShortPassing','Volleys',
             'Dribbling','Curve','FKAccuracy','LongPassing','BallControl',
             'Acceleration','SprintSpeed','Agility','Reactions','Balance',
             'ShotPower','Jumping','Stamina','Strength','LongShots',
             'Aggression','Interceptions','Positioning','Vision','Penalties',
             'Composure','Marking','StandingTackle','SlidingTackle']
can_abilities = candidate[abilities]
can_infor = candidate[infor]
can_new = pd.merge (can_infor,can_abilities,left_index=True,right_index=True)
can_new.head()
```

```python
CR = candidate[abilities] [candidate['Name'] == 'Cristiano Ronaldo']
CR.head()
```

```python
categories=list(CR)[0:]
N = len(categories)
```

```python
values=CR.loc[1].values.flatten().tolist()
values += values[:1]
```

```python
from math import pi
import matplotlib.pyplot as plt
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
```

```python
plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], categories)
ax.plot(angles, values, linewidth=1, linestyle='solid')
ax.fill(angles, values, 'blue', alpha=0.1)
plt.show()
```

```python
from sklearn.preprocessing import scale 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, LassoCV
from sklearn.metrics import mean_squared_error
from sklearn import preprocessing
from sklearn.linear_model import LassoLarsCV
```

```python
y = can_new['Overall']
X = can_new[abilities]
alphas = 10**np.linspace(10,-2,100)*0.5
```

```python
X_train, X_test , y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=1)
```

```python
model=LassoLarsCV(cv=10, precompute=False).fit(X_train,y_train)
```

```python
lasso = Lasso(max_iter = 10000, normalize = True)
coefs = []

for a in alphas:
    lasso.set_params(alpha=a)
    lasso.fit(scale(X_train), y_train)
    coefs.append(lasso.coef_)
    
```

```python
lassocv = LassoCV(alphas = None, cv = 500, max_iter = 100000, normalize = True)
lassocv.fit(X_train, y_train)
```
